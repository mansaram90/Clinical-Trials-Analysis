# Pharma Data Analysis Portfolio Project (Python, pandas)

This project showcases end-to-end **data analysis in Python** for the pharmaceutical domain using multiple interrelated datasets:

1. **Clinical Trials** (`clinical_trials.csv`)  
2. **Adverse Events** (`adverse_events.csv`)  
3. **FDA Drug Approvals** (`fda_approvals.csv`)  
4. **Pharma Sales** (`pharma_sales.csv`)  

The goal is to analyze these datasets individually and together, calculate **key KPIs**, and generate insights similar to what real pharma companies do for R&D, safety, approvals, and commercial performance.



## 1. Problem Statement

Pharmaceutical companies generate and consume huge amounts of data across the drug lifecycle:

- **R&D / Clinical Trials**: Which diseases and phases get the most focus? How long do trials take?  
- **Safety / Pharmacovigilance**: Which drugs have more serious adverse events?  
- **Regulatory Approvals**: How many drugs get approved per year? Which companies lead?  
- **Commercial / Sales**: How do sales and marketing spend vary by region and product?  

This project aims to:

1. Perform **exploratory data analysis (EDA)** on each dataset.  
2. Compute **business-relevant KPIs** in the pharma context.  
3. Create **visual summaries** of important trends.  
4. Demonstrate how to **link datasets** (e.g., Drug → Approvals → Sales → Safety).  

All analysis is done in **Python using pandas and matplotlib**, without machine learning, to build a strong foundation in data analysis.


```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 120)
```

## 2. Load Datasets

```python

clinical_trials_path = "clinical_trials.csv"
adverse_events_path = "adverse_events.csv"
fda_approvals_path = "fda_approvals.csv"
pharma_sales_path = "pharma_sales.csv"

df_trials = pd.read_csv(clinical_trials_path)
df_ae = pd.read_csv(adverse_events_path)
df_approvals = pd.read_csv(fda_approvals_path)
df_sales = pd.read_csv(pharma_sales_path)

df_trials.head()
```

### Quick look at each dataset

```python

print("Clinical Trials:")
display(df_trials.head())

print("\nAdverse Events:")
display(df_ae.head())

print("\nFDA Approvals:")
display(df_approvals.head())

print("\nPharma Sales:")
display(df_sales.head())
```


## 3. Clinical Trials Analysis

### 3.1 KPIs for Clinical Trials

We will compute:
- Trials per phase  
- Trials per condition (disease)  
- Median and mean enrollment size    
- Trials per country  
- Trials per sponsor  


```python

# Convert dates and compute duration
df_trials['StartDate'] = pd.to_datetime(df_trials['StartDate'], errors='coerce')
df_trials['PrimaryCompletionDate'] = pd.to_datetime(df_trials['PrimaryCompletionDate'], errors='coerce')

df_trials['Duration_Days'] = (df_trials['PrimaryCompletionDate'] - df_trials['StartDate']).dt.days

# Ensure EnrollmentCount is numeric
df_trials['EnrollmentCount'] = pd.to_numeric(df_trials['EnrollmentCount'], errors='coerce')

df_trials[['StartDate', 'PrimaryCompletionDate', 'Duration_Days', 'EnrollmentCount']].describe()
```

```python

trials_per_phase = df_trials['Phase'].value_counts().sort_index()
print(trials_per_phase)

plt.figure(figsize=(6,4))
trials_per_phase.plot(kind='bar')
plt.title("Number of Trials per Phase")
plt.xlabel("Phase")
plt.ylabel("Number of Trials")
plt.tight_layout()
plt.show()
```

```python

trials_per_condition = df_trials['Condition'].value_counts()
print(trials_per_condition.head(10))

plt.figure(figsize=(8,5))
trials_per_condition.head(10).plot(kind='bar')
plt.title("Top 10 Conditions by Number of Trials")
plt.xlabel("Condition")
plt.ylabel("Number of Trials")
plt.tight_layout()
plt.show()
```

```python

enrollment_stats = df_trials['EnrollmentCount'].describe()
enrollment_stats
```

```python

plt.figure(figsize=(8,5))
df_trials['EnrollmentCount'].plot(kind='hist', bins=40)
plt.title("Enrollment Count Distribution")
plt.xlabel("Enrollment Count")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python

duration_stats = df_trials['Duration_Days'].describe()
duration_stats
```

```python

trials_per_country = df_trials['LocationCountry'].value_counts()
print(trials_per_country)

plt.figure(figsize=(8,5))
trials_per_country.plot(kind='bar')
plt.title("Trials per Country")
plt.xlabel("Country")
plt.ylabel("Number of Trials")
plt.tight_layout()
plt.show()
```


## 4. Adverse Events (Safety) Analysis

### 4.1 KPIs for Adverse Events

Using `adverse_events.csv`, we analyze:
- Number of reports per drug  
- Most common reactions  
- Serious vs non-serious proportions  
- Age and gender distribution of adverse events  


```python

df_ae['ReportDate'] = pd.to_datetime(df_ae['ReportDate'], errors='coerce')

print(df_ae.describe(include='all'))
```

```python

ae_per_drug = df_ae['DrugName'].value_counts()
print(ae_per_drug)

plt.figure(figsize=(6,4))
ae_per_drug.plot(kind='bar')
plt.title("Number of Adverse Event Reports per Drug")
plt.xlabel("Drug")
plt.ylabel("Number of Reports")
plt.tight_layout()
plt.show()
```

```python

reactions = df_ae['Reaction'].value_counts()
print(reactions)

plt.figure(figsize=(8,5))
reactions.plot(kind='bar')
plt.title("Most Common Reactions")
plt.xlabel("Reaction")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python

serious_counts = df_ae['Serious'].value_counts()
serious_counts.index = serious_counts.index.map({0: "Non-Serious", 1: "Serious"})

plt.figure(figsize=(4,5))
ax = serious_counts.plot(kind='bar')

plt.title("Serious vs Non-Serious Adverse Events")
plt.xlabel("Event Type")
plt.ylabel("Count")

# Add count labels above bars
for container in ax.containers:
    ax.bar_label(container)

plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12,5))

# --- Age Histogram + KDE ---
plt.subplot(1, 2, 1)
sns.histplot(df_ae['Age'], bins=20, kde=True, edgecolor='black')
plt.title("Age Distribution (with KDE)")
plt.xlabel("Age")
plt.ylabel("Count")

# --- Gender Distribution ---
plt.subplot(1, 2, 2)
sns.countplot(x=df_ae['Gender'])
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```


## 5. FDA Drug Approvals Analysis

### 5.1 KPIs for FDA Approvals

Using `fda_approvals.csv`, we analyze:
- Approvals per year  
- Approvals per company  
- Small molecules vs biologics  
- Approvals per indication (disease area)  


```python

print(df_approvals.describe(include='all'))
```

```python

approvals_per_year = df_approvals['ApprovalYear'].value_counts().sort_index()
print(approvals_per_year)

plt.figure(figsize=(8,5))
approvals_per_year.plot(kind='bar')
plt.title("Number of Drug Approvals per Year")
plt.xlabel("Year")
plt.ylabel("Number of Approvals")
plt.tight_layout()
plt.show()
```

```python

approvals_by_type = df_approvals['Type'].value_counts()
print(approvals_by_type)

plt.figure(figsize=(5,4))
approvals_by_type.plot(kind='bar')
plt.title("Approvals by Type (Small Molecule vs Biologic)")
plt.xlabel("Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python

approvals_by_company = df_approvals['Company'].value_counts()
print(approvals_by_company)

plt.figure(figsize=(6,4))
approvals_by_company.plot(kind='bar')
plt.title("Approvals by Company")
plt.xlabel("Company")
plt.ylabel("Number of Approvals")
plt.tight_layout()
plt.show()
```


## 6. Pharma Sales Analysis

### 6.1 KPIs for Sales

Using `pharma_sales.csv`, we analyze:
- Revenue per region  
- Units sold per product  
- Monthly revenue trend  
- Marketing spend vs revenue (simple ratio)  


```python

df_sales['Month'] = pd.to_datetime(df_sales['Month'], errors='coerce')

print(df_sales.describe(include='all'))
```

```python

revenue_by_region = df_sales.groupby('Region')['RevenueUSD'].sum().sort_values(ascending=False)
print(revenue_by_region)

plt.figure(figsize=(6,4))
revenue_by_region.plot(kind='bar')
plt.title("Total Revenue by Region")
plt.xlabel("Region")
plt.ylabel("Revenue (USD)")
plt.tight_layout()
plt.show()
```

```python

units_by_product = df_sales.groupby('Product')['UnitsSold'].sum().sort_values(ascending=False)
print(units_by_product)

plt.figure(figsize=(6,4))
units_by_product.plot(kind='bar')
plt.title("Total Units Sold by Product")
plt.xlabel("Product")
plt.ylabel("Units Sold")
plt.tight_layout()
plt.show()
```

```python

monthly_revenue = df_sales.groupby(df_sales['Month'].dt.to_period('M'))['RevenueUSD'].sum()
print(monthly_revenue.head())

plt.figure(figsize=(10,4))
monthly_revenue.plot()
plt.title("Monthly Revenue Over Time")
plt.xlabel("Month")
plt.ylabel("Revenue (USD)")
plt.tight_layout()
plt.show()
```

```python

df_sales['MarketingROI'] = df_sales['RevenueUSD'] / df_sales['MarketingSpend']

roi_summary = df_sales['MarketingROI'].describe()
roi_summary
```


## 7. Linking Datasets (Integration Example)

Here we demonstrate how to link:

- **FDA Approvals** - **Sales** by `DrugName` / `Product`  
- **FDA Approvals** - **Adverse Events** by `DrugName`  

This simulates an integrated view from **approval → safety → commercial performance**.


```python

# Example: link approvals to sales via DrugName/Product
merged_approvals_sales = df_approvals.merge(
    df_sales,
    left_on='DrugName',
    right_on='Product',
    how='inner'
)

# merge AEs 
merged_all = merged_approvals_sales.merge(
    df_ae,
    left_on='DrugName',
    right_on='DrugName',
    how='left'
)

merged_all.head()
```

```python
df_ae.head()
```

```python

# Aggregate revenue by DrugName from the merged data
revenue_by_drug = merged_approvals_sales.groupby('DrugName')['RevenueUSD'].sum().sort_values(ascending=False)
print(revenue_by_drug)

plt.figure(figsize=(6,4))
revenue_by_drug.plot(kind='bar')
plt.title("Total Revenue for Approved Drugs")
plt.xlabel("Drug Name")
plt.ylabel("Revenue (USD)")
plt.tight_layout()
plt.show()
```

```python

# Example: link approvals to adverse events via DrugName
merged_approvals_ae = df_approvals.merge(
    df_ae,
    left_on='DrugName',
    right_on='DrugName',
    how='inner'
)

merged_approvals_ae.head()
```

```python

# Compute serious AE rate per DrugName for approved drugs
serious_rate_by_drug = merged_approvals_ae.groupby('DrugName')['Serious'].mean()
serious_rate_by_drug
```
